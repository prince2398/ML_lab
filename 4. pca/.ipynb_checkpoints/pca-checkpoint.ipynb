#### AIM : Principal Component Analysis
##### To Perform PCA on given dataset and reduce its dimensionity. 
#### DESCRIPTION:
Principal component analysis (PCA) is a statistical procedure that uses an orthogonal transformation to convert a set of observations of possibly correlated variables (entities each of which takes on various numerical values) into a set of values of linearly uncorrelated variables called principal components. 
In Machine Learning, this is used to reduce dimentionality of data. It is useful because it makes data simpler and more interpretable and visualization of data is also easy. In this, dimensions are reduced without much loss of data.
##### Steps in PCA : 
1. Take whole dataset consisting of d-dimensional samples ignoring class label
2. Compute d-dimensional mean vector
3. Computer Covariance Matrix
4. Compute Eigen Vectors and corresponding Eigen Values
5. Sort Eigen Vectors with decreasing Eigen Values 6. Choose k Eigen Vectors with largest Eigen Values to form d*k dimensional matrix (where every column represents an eigen vector
7. Use this d*k matrix to transform d dimensional data to k dimensions 
#### CODE and OUTPUT : 

```python
# Principal Component Analysis 
# Libraries
import numpy as np 
import scipy as sp
from sklearn import datasets
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# We are using IRIS dataset provided by sklearn
iris = datasets.load_iris()
X = iris.data
y = iris.target
print(X.shape,y.shape)
```

```python
# Calculation of Mean Vector
meanVector = np.mean(X,axis=0)
print(meanVector) 
```

```python
# Calculation of Covariance matrix
Xnew = X - meanVector
cov = (np.matmul(np.transpose(Xnew),Xnew)) / Xnew.shape[0]
print(cov) 
```

```python
# Calculating Eigen Values and Eigen Vectors
eVal, eVec = np.linalg.eig(cov)
```

```python
# Sorting Eigen Vectors according to decreasing Eigen Values
idx = eVal.argsort()[::-1]
eValSorted = eVal[idx]
eVecSorted = eVec[:, idx]
print(eValSorted)
print(eVecSorted) 
```

```python
# Selecting Principal Components
pComps = eVecSorted[:, :2]
print(pComps) 
```

```python
# Transforming sample data to principal components
PCAdata = np.matmul(X,pComps)
print(PCAdata.shape)
```

```python
# Representation  of PCA Data
plt.scatter(PCAdata[:,0],PCAdata[:,1],c=y)
plt.title('Principal Component Analysis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Compoment 2')
plt.show()
```
